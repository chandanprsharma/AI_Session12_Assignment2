#### This assignment is for visualization using matplotlib:
**data to use:**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head(2)
```

#### Charts to plot:

#### Q1. Create a pie chart presenting the male/female proportion

```python
m_f_prop = titanic.groupby('sex').count().reset_index()
plt.pie(m_f_prop.pclass, labels=m_f_prop.sex, autopct='%1.0f%%', shadow=True, explode=[.1, 0]);
plt.axis("equal");
plt.show();
```

#### Q2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
# Generate Data
x = titanic.fare
y = titanic.age
labels = titanic.sex
df = pd.DataFrame(dict(x=x, y=y, label=labels))

groups = df.groupby('label')

# Plot
fig, ax = plt.subplots()
ax.margins(0.02) # Optional, just adds 2% padding to the autoscaling
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5, label=name)
ax.legend()
plt.title('Scatterplot with the Fare paid and the Age')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.gcf().set_size_inches((15, 8))
plt.show()
```
