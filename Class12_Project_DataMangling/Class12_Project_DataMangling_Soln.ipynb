#### Q18. Determine how much quantity of each product was purchased by each user
**Expected Output:**


```python
users['Key'] = 0
products['Key'] = 0
users_products_cross = pd.merge(users, products, how='outer', left_on='Key', right_on='Key')
u_p_t_df = pd.merge(users_products_cross, transactions, how='outer', left_on=["UserID","ProductID"], right_on=["UserID","ProductID"])
u_p_t_df.drop(['User', 'Gender', 'Registered', 'Cancelled', 'Key', 'Product', 'Price', 'TransactionID', 'TransactionDate'], axis=1, inplace=True)
u_p_t_df['Quantity'].fillna(value=0, inplace=True)
u_p_t_df = u_p_t_df.groupby(['UserID', 'ProductID']).agg({'Quantity':'sum'}).reset_index()
u_p_t_df = u_p_t_df.astype({"UserID": int})
u_p_t_df.head(15)
```

```python
data = pd.merge(users, transactions, how='left', on='UserID').head(10)
data.drop(columns=['Key'], inplace=True)
```

```python
my_columns = list(data.columns)
my_columns
```

```python
#set threshold to drop NAs
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)
```

```python
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0] 
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
#count of missing data
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0] 
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
