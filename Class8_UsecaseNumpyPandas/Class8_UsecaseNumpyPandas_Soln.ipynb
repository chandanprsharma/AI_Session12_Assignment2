#### Q1) Write a Python program which accepts a list named : randomList = ['a', 0, 2]. Use exception handling using try-catch which gives the output as:



```python
# function to divide two numbers
def findRecipocal(rl):
    for i in rl:
        try:
            recipocal = 1/i
            print(f'The entry is {i} \nThe recipocal of {i} is {recipocal}')
        except TypeError as te:
            # handle ValueError exception
            print("Oops! <class '\nValueError'> occured Next entry.")
        except ZeroDivisionError as zde:
            # handle ZeroDivisionError
            print("Oops! <class '\nZeroDivisionError'> occured Next entry.")

randomList = ['a', 0, 2]
findRecipocal(randomList)
```

#### Q2) Array out of Bound Exception

Write a Python program to give exception Array Out of Bound if the user wants to access the elements beyond the list size (use try and except)

```python
# function to access list element of given index
def accessList(l, index):
    try:
        print(f'The number at index {index} is {l[index]}')
    except IndexError as ie:
        # handle IndexError exception
        print("Exception: Array Out of Bound")

lst = list(range(1,11))
index = int(input('Enter the index of element you want to access: '))
accessList(lst, index)
```

#### Q5) Write a program in Python with one class called Cipher. Within the constructor of this class, ask user for a string and store it. Use a static variable, key to store a randomly generated integer between 1 and 50 inclusive. Implement two methods, encrypt and decrypt within this class. Encrypt generates and prints a cipher text using the user-entered string and the key and ecrypt generates decrypted string from ciphertext. The cipher only encrypts alpha and numeric (A-Z, a-z, 0-9). All Symbols, such as - , ; %, remain unencrypted. The cipher text can have special characters. Use generator expression to filter out alpha and numeric characters of the input string and to generate cipher text. Create an instance of this class, encrypt and decrypt back the user entered string.

```python
import random
```

```python
class Cipher:

    key = random.randint(1,51)
    
    def __init__(self, text):
        self.text = text
        self.encrypted_text = ''
        
    def filterAlNum(self):
        for c in self.text:
            yield c.isalnum()

    def encrypt(self):
        isAlNum = self.filterAlNum()
            
        for c in self.text:
            if next(isAlNum):
                self.encrypted_text += chr(ord(c) + Cipher.key)
            else:
                self.encrypted_text += c
        print('Encrypted cipher text: ', self.encrypted_text)

        
    def decrypt(self):
        decrypted_text = ''
            
        for index, value in enumerate(self.text):
            if value.isalnum():
                decrypted_text += chr(ord(self.encrypted_text[index]) - Cipher.key)
            else:
                decrypted_text += self.encrypted_text[index]
                
        print('Decrypted cipher text: ', decrypted_text)
        
        
        

msg = 'Hello128#$1'
obj = Cipher(msg)
obj.encrypt()
obj.decrypt()
```
